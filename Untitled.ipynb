```python
class MorpionEnv:
    def __init__(self):
        # Les 8 combinaisons gagnantes pré-calculées (index 0 en bas à gauche à 8 en haut à droite)
        self.WIN_MASKS = [7, 56, 448, 73, 146, 292, 273, 84]
        self.reset()

    def reset(self):
        """Remet le plateau à zéro."""
        self.board_X = 0  # Bitboard du Joueur X
        self.board_O = 0  # Bitboard du Joueur O
        return self.board_X, self.board_O

    def get_legal_moves(self):
        """Retourne la liste des cases vides (index de 0 à 8)."""
        # OU logique pour l'occupation
        occupation = self.board_X | self.board_O
        # NON logique (~) pour les cases vides, et ET (&) 511 pour nettoyer au-delà du 9ème bit
        vides_bits = (~occupation) & 511
        
        # On convertit le masque de bits en liste d'index (ex: [0, 2, 4])
        legal_moves = []
        for i in range(9):
            if (vides_bits >> i) & 1:
                legal_moves.append(i)
        return legal_moves

    def step(self, action, player):
        """Joue un coup pour un joueur (player = 'X' ou 'O')."""
        move_mask = 1 << action
        if player == 'X':
            self.board_X |= move_mask
        else:
            self.board_O |= move_mask

    def check_win(self, player_board):
        """Vérifie si le bitboard donné contient une victoire."""
        for mask in self.WIN_MASKS:
            if (player_board & mask) == mask:
                return True
        return False

    def is_draw(self):
        """Vérifie si le plateau est plein (511 = 9 bits à 1)."""
        return (self.board_X | self.board_O) == 511
```
